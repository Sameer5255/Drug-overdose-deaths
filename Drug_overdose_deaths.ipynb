## Handling Nan Null value

```python
#data information
data.info()
```

```python
columns_to_fill = ['DateType', 'Age']
data[columns_to_fill] = data[columns_to_fill].apply(lambda col: col.fillna(col.mean()))
```

```python
data.isnull().sum()
```

```python
data.replace('null',np.nan,inplace=True)
```

```python
data.isnull().sum()
```

```python
# Replace 'null' strings with np.nan (you've already done this)
data.replace('null', np.nan, inplace=True)

# Columns to fill with mode
columns_to_fill = ['Race', 'Sex', 'ResidenceCity', 'ResidenceCounty','ResidenceState','DeathCity','DeathCounty','Location','LocationifOther','DescriptionofInjury','InjuryPlace','InjuryCity','InjuryCounty','InjuryState','OtherSignifican','MannerofDeath','ResidenceCityGeo','InjuryCityGeo']

# Fill NaN values in those columns with mode (most frequent value)
for col in columns_to_fill:
    if not data[col].mode().empty:  # Check to avoid errors if column is entirely NaN
        data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
data.drop('Other', axis=1, inplace=True)
```

## Checking which drug has highest number of

```python
# List of drug columns (from your dataset)
drug_columns = [
    'Heroin', 'Cocaine', 'Fentanyl', 'Fentanyl_Analogue', 'Oxycodone',
    'Oxymorphone', 'Ethanol', 'Hydrocodone', 'Benzodiazepine',
    'Methadone', 'Amphet', 'Tramad', 'Morphine_NotHeroin',
    'Hydromorphone', 'OpiateNOS', 'AnyOpioid'
]

# Count how many deaths involved each drug (i.e., count non-zero entries)
drug_death_counts = data[drug_columns].apply(pd.to_numeric, errors='coerce').sum().sort_values(ascending=False)

# Create a DataFrame for plotting
drug_deaths_df = drug_death_counts.reset_index()
drug_deaths_df.columns = ['Drug', 'Death Count']

# Plot using Plotly
fig = px.bar(drug_deaths_df, x='Drug', y='Death Count',
             title='Number of Deaths Involving Each Drug',
             text='Death Count',
             labels={'Death Count': 'Number of Deaths'},
             color='Death Count',
             color_continuous_scale='Reds')

fig.update_traces(textposition='outside')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

Heroin has highest number of death followed by opiods.


Other depressants like Benzodiazepines and Alcohol (Ethanol) also play significant roles.

This highlights the importance of addressing opioid and polydrug use in prevention efforts.

The data suggests targeted interventions around opioid abuse could have a strong impact on reducing deaths.

## Check number of patient taking number of drugs

```python
# List of drug-related columns
drug_columns = [
    'Heroin', 'Cocaine', 'Fentanyl', 'Fentanyl_Analogue', 'Oxycodone',
    'Oxymorphone', 'Ethanol', 'Hydrocodone', 'Benzodiazepine',
    'Methadone', 'Amphet', 'Tramad', 'Morphine_NotHeroin',
    'Hydromorphone', 'OpiateNOS', 'AnyOpioid'
]

# Convert drug columns to numeric in case of strings like '1', '0', 'NaN'
drug_data = data[drug_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Count number of drugs taken per patient (i.e., sum across the row)
data['Num_Drugs_Taken'] = drug_data.sum(axis=1)

# Count how many patients took each number of drugs
drug_counts = data['Num_Drugs_Taken'].value_counts().sort_index()

# Prepare DataFrame for plotting
df_plot = drug_counts.reset_index()
df_plot.columns = ['Number of Drugs', 'Number of Patients']

# Plot
fig = px.bar(df_plot,
             x='Number of Drugs',
             y='Number of Patients',
             title='Number of Patients vs Number of Drugs Taken',
             text='Number of Patients')

fig.update_traces(textposition='outside')
fig.update_layout(xaxis=dict(dtick=1), yaxis_title='Patients')

fig.show()
```

64 patients not taking any drug.

946 patients taking single drug.

1476 patients taking 2 drugs which is higest in patient number .

Only 1 patient is taking 8 drugs.


The majority of patients involved in this dataset tend to have taken multiple drugs, often 2 or 3 drugs.

Polydrug use (taking multiple drugs) is common and may have a strong relationship with patient outcomes.

Cases with very high numbers of drugs taken (6-8) are rare but still present.

This pattern could indicate a need for targeted interventions addressing polydrug use risks.

##Checking which Country has more Accident type of Death  

```python


# Filter for accidental deaths
accident_data = data[data['MannerofDeath'] == 'Accident']

# Count deaths by ResidenceCounty
accident_counts = accident_data['ResidenceCounty'].value_counts().nlargest(10).reset_index()
accident_counts.columns = ['ResidenceCounty', 'Death Count']

# Plot top 10 counties
fig = px.bar(accident_counts,
             x='ResidenceCounty',
             y='Death Count',
             title='Top 10 Residence Counties by Number of Accidental Deaths',
             labels={'ResidenceCounty': 'Residence County', 'Death Count': 'Number of Deaths'})

fig.update_layout(xaxis_tickangle=-45, margin=dict(b=150))
fig.show()
```

Hartford County has the highest number of accidental deaths by a significant margin, with around 2000 deaths.

New Haven County follows as the second highest, but with just over half the deaths of Hartford (~1100 deaths).

The numbers then steadily decline through Fairfield, New London, Litchfield, Middlesex, Windham, Tolland, Westchester, and Dutchess counties.

Westchester and Dutchess counties have the lowest counts among the top 10, close to zero or very minimal.

The large gap between Hartford and the other counties might indicate:

A higher population in Hartford

Possible socio-economic or environmental factors increasing accidental death risks in Hartford

Greater reporting or data availability in Hartford County
